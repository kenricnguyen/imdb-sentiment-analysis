```python
from keras.preprocessing import sequence
from keras.models import Sequential, model_from_json
from keras.layers import Dense, Embedding,Flatten, Activation
from keras.datasets import imdb

import numpy
import os
```

```python
# loading the top 20000 words from imdb reviews
(X_train,y_train),(X_test,y_test) = imdb.load_data(path="imdb.npz",num_words=5000)
```

```python
# limit to the first 100 words for each review
X_train = sequence.pad_sequences(X_train, maxlen=100)
X_test = sequence.pad_sequences(X_test , maxlen=100)
```

```python
# build a NN model of 2 layers: Word embedding layer and a relu activation layer
model = Sequential()
model.add(Embedding(5000,32,input_length=100))
model.add(Flatten())
model.add(Activation('relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
# compile the model using adam optimizer
model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, batch_size=32, epochs=5, verbose=2, validation_data =(X_test, y_test))
```

```python
score, accuracy = model.evaluate(X_test, y_test, batch_size=32, verbose=2)

print('Score:',score)
print('Accuracy:',accuracy)
```

```python
print("Saving model and weights to disk ...")
# save model to JSON
model_json = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)
    
# save weights to hdf5
model.save_weights("model_weights.h5")

print("Finished !")
```

```python
print("Loading model from disk ...")
# load model from JSON file
json_file = open('model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)

#load weights into new model
loaded_model.load_weights("model_weights.h5")
print("Done loading model!")
```

```python
# compile loaded_model
loaded_model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])

# evaluate loaded_model on test data
score, accuracy = loaded_model.evaluate(X_test, y_test, batch_size=32, verbose=2)

print('Score:',score)
print('Accuracy:',accuracy)
```

```python
import matplotlib.pyplot as plt
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
